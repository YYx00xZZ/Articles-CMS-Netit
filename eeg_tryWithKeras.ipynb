```
# alg. 1
def filter_ids(data):
    EventID_NEW = ''
    EventID_TO_SET = ''
    for index, row in data.iterrows(): #итерираме ред по ред с вградения метод iterrows() от pandas
        # достъпваме данни чрез име на колона
        EventID_NEW = str(row['EventId']) # prochetenoto sudurjanie na suotvetnata kletka
        if EventID_NEW == '0':
            data.at[index, 'EventId'] = str(EventID_TO_SET)
        else:
            EventID_NEW = str(row['EventId'])
            EventID_TO_SET = str(row['EventId'])
    return data


# alg. 2
def filter_events(data):
    """
    Функцията очаква pandas.DataFrame;
    Връща DataFrame съдържащ данни само за ляво и дясно
    """
    data = data[(data.EventId == '33025') | (data.EventId == '33026')]
    return data
```

```
import pandas as pd
# first neural network with keras tutorial
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense

input_file_url = 'https://github.com/YYx00xZZ/UniBIT-Study/blob/colab/data/input/1_GD_Standart_14ch_24.01.2020.csv?raw=true'
initial_col_names = ['Time128Hz', 'Epoch', 'AF3', 'F7', 'F3', 'F5', 'T7', 'P7', 'O1', 'O2', 'P8', 'T8', 'FC6', 'F4', 'F8', 'AF4', 'EventId', 'EventDate', 'EventDuration']
df = pd.read_csv(input_file_url, names=initial_col_names, index_col=False, header=None, sep=',',skiprows=1, dtype={'EventId' : str})
df.fillna(0, inplace=True)

df = filter_events(filter_ids(df))


# side_effects_df = df.drop(['Epoch','Time128Hz', 'EventId','EventDate', 'EventDuration'], axis=1)
# (
#     side_effects_df
#                     .reset_index()
#                     .drop(['index'], axis=1)
# )
```

```
dataset = loadtxt('/content/train_test_set.csv', skiprows=1, delimiter=",")
```

```
# split into input (X) and output (y) variables
# X = data.iloc[:, 0:-1]
# y = data.iloc[:, -1]
X = dataset[:,0:-1]
y = dataset[:,-1]
```

```
# define the keras model
model = Sequential()
model.add(Dense(2, input_dim=14, activation='sigmoid'))
# model.add(Dense(7, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```
import keras
model = keras.Sequential([keras.layers.ReLU(max_value=None, negative_slope=5400.0, threshold=0.0),
                keras.layers.Dense(128, input_dim=14, activation = 'relu'),                          
                keras.layers.Dense(1,activation = 'softmax')])
```

```
# compile the keras model
# keras.layers.ReLU(max_value=None, negative_slope=0.0, threshold=0.0)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(X, y, epochs=150, batch_size=32)
```

```
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```
# make class predictions with the model
predictions = model.predict_classes(X)
# summarize the first 5 cases
for i in range(1000):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```
